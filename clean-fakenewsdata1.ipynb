```python
path = 'datasets/fakenewsdata1/PublicData'
```

```python
#creat empty df
df = pd.DataFrame(columns = ['content', 'label'])
wanted = ['Fake', 'Real']

#loop over all the directories
for folder in os.listdir(path):
    if folder == 'README.txt':
        continue
    tmp_path = os.path.join(path,folder)
    for fake_real in os.listdir(tmp_path):
        if fake_real in wanted:
            complete_path = os.path.join(tmp_path, fake_real)
            print('Reading data from: ', complete_path)
            for text in os.listdir(complete_path):
                to_load = os.path.join(complete_path, text)
                
                #if we are in a real directory
                if fake_real == 'Real':
                    
                    #load the text with pandas
                    tmp = pd.read_csv(to_load,
                                        encoding='cp1252',
                                        error_bad_lines=False, 
                                        header=None,
                                        delimiter='\n')

                    
                    #and change it so, we can actually use it afterwards
                    data = tmp.apply(' '.join).reset_index()
                    del data['index']
                    data['label'] = 0
                    data.columns=['content','label']
                    df = df.append(data)

                    
                #Do the same for fake directories
                elif fake_real == 'Fake':
                    tmp = pd.read_csv(to_load,
                                        encoding='cp1252',
                                        error_bad_lines=False, 
                                        header=None,
                                        delimiter='\n')

                    data = tmp.apply(' '.join).reset_index()
                    del data['index']
                    data['label'] = 1
                    data.columns=['content','label']
                    df = df.append(data)
```

```python
#Reindex, to have them indexed. lul
df.reset_index()
#del df['index']
```
